### 2. Entendimiento de los Datos

```python
# Recolección de los datos

import csv

with open("communities.data", "r") as file:
    reader = csv.reader(file)
    data = [row for row in reader]

data = pd.DataFrame(data)
data
```

```python
# Imprime todo el contenido de communities
with open("communities.names", "r") as file:
    for line in file:
        print(line.strip())
```

```python
# Imprime solo las variables con su descripcion
count=0

with open("communities.names", "r") as file:
    for line in file:
        if line.strip():  # Verificar que la línea no esté vacía
            if line.strip()[0:3] == '-- ' and (line.strip()[-1] == ')' or line.strip()[-1] == '#'):
                print(f'Column No. {count}:', line.strip()[2:])
                count+=1
```

```python
# Revisemos cuantos valores faltantes (?) tiene la columna 1
print("====== Valores Faltantes ======")
print(data[[1]].value_counts())
print()
# Valores faltantes (?) en la columna 1
print("====== En Porcentaje ======")
print(((data[[1]] == '?').sum() / data[[1]].value_counts().sum()) * 100)
```

Para todas las columnas que tengan un nivel porcentaje mayor al 50% de valores faltantes podemos eliminarlas debido a que son demasiados datos.

```python
(((data[127] == '?').sum() / data[127].value_counts().sum()) * 100)
```

```python
# Hacemos una lista de las columnas con valores faltantes por encima del 50%
cols = []
for column in data.columns:
    if (((data[column] == '?').sum() / data[column].value_counts().sum()) * 100) > 50:
        cols.append(column)
print('Total de Columnas:', len(cols))
print(cols)
```

```python
# Eliminamos columnas
data = data.drop(cols, axis=1)
```
